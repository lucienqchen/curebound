```python
# Readjust the clinical trials name columns to plot in Tableau

def adjust_cli_trials_names(data, title):

    """
        df (DataFrame): subject dataframe
        title (str): the title of the project to adjust clinical trials names

    """

    df = data[data['Project Title'] == title].reset_index(drop=True)
    num_trials = df['Clinical Trials Resulted'].sum()
    first = df['First Clinical Trial'].reset_index(drop=True)
    second = df['Second Clinical Trial'].reset_index(drop=True)

    if num_trials <= 1:
        return df

    while len(first[~first.isna()]) > 1:
        first_nonnull_index = first[~first.isna()].index[-1]
        to_move = first.iloc[first_nonnull_index]
        first_null_index = second[second.isna()].index[0]
        second.iloc[first_null_index] = to_move
        first[first_nonnull_index] = np.NaN

    df['First Clinical Trial'] = first
    df['Second Clinical Trial'] = second
    return df
```
