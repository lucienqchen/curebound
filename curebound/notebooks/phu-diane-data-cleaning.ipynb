```python
# International	 National   Local	General Public	Heathcare  Professionals  Patients	Patient Families

def impute_bool(df, col):

    # Replace Checked/Unchecked with True/False

    df[col] = df[col].replace({'Checked': True, 'Unchecked': False})
    return None

for col in pr.columns:
    if 'International' in col or 'National' in col or 'Local' in col \
        or 'General Public' in col or 'Healthcare Professionals' in col \
        or 'Patients' in col or 'Patient Families' in col:
        impute_bool(pr, col)

pr.head()
```
